## 0.6 : Write/read a text file

We review how I/O operations with text files work in Python.

```python
# open the file hello.dat in write mode
with open('hello.dat', 'w') as text_file :
    text_file.write('Hello!!') # write
```

```python
# give a quick look at the file
!cat hello.dat
```

```python
# open the file hello.dat in read mode
with open('hello.dat', 'r') as text_file :
    data = text_file.read() # read

# data holds the content read from the file
print(data) # show
```

## 0.7 : Convert Python data structures to/from a JSON file

We show how Python data structures can be easily dumped in/loaded from a JSON file.

Let's create a Python data structure, similar to what we will be using in the next tutorial.

```python
# a dictionary contains keys and values
data={}
data['project']='liquid_project'

# it can be nested
data['material']={}
data['material']['name']='water'
data['material']['formula']='H2O'
data['material']['density']=1

# lists contain information which can be iterated
data['material']['atomic_species']=['O','H']

# show the data
print(data)
```

Write the Python data in the JSON file `data.json`.

```python
import json

# write data to JSON file
with open('data.json', 'w') as file:
    json.dump(data, file)
```

Give a quick look at the file `data.json`.

```python
# give a quick look at the file
!cat data.json
```

Read the Python data from the JSON file `data.json`.

```python
import json

# read data from JSON file
with open('data.json', 'r') as file:
    d = json.load(file)

# show the data
print(d)

# pretty print the data
print(json.dumps(d, indent=2))
```

## 0.8 : Convert Python data structures to/from a YAML file

We show how Python data structures can be easily dumped in/loaded from a YAML file.

Write the Python data in the YAML file `data.yml`.

```python
import yaml

# write data to YAML file
with open('data.yml', 'w') as file:
    yaml.dump(data, file, sort_keys=False)
```

Give a quick look at the file `data.yml`.

```python
# give a quick look at the file
!cat data.yml
```

Read the Python data from the YAML file `data.yml`.

```python
import yaml

# read data from YAML file
with open('data.yml', 'r') as file:
    d = yaml.load(file, Loader=yaml.FullLoader)

# show the data
print(d)

# pretty print the data
print(yaml.dump(d, sort_keys=False))
```

## 0.9 : Generate data and make a simple plot

We are going to generate the (X,Y) data points for the functions: $Y=X$, and $Y=X^2$, then plot the data points.

```python
import numpy as np

# generate 6 points equally spaced from 0 to 3.
x = np.linspace(0,3,6)

# generate y points for each x (implicit iteration)
y1 = x
y2 = x*x

# show data
print('x  : ', x)
print('y1 : ', y1)
print('y2 : ', y2)
```

```python
import matplotlib.pyplot as plt

# plot 
fig, ax = plt.subplots(1, 1)
ax.plot(x, y1, 'bo-', label='$y=x$' )
ax.plot(x, y2, 'ro-', label='$y=x^2$' )

# set labels 
plt.xlabel('x')
plt.ylabel('y')

# set legend
plt.legend()

# set title 
plt.title('Functions')

# show the plot
plt.show()
```
